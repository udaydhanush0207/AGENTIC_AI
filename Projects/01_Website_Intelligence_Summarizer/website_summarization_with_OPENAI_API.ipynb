```python
system_prompt = """
You are a summary assitant for a comapny website as well as its articles and all sometimes you will be prvoided with acrnoyms 
related to company and then you need to explain those as well with proper bulet points and sub headers and then think of someone who is 
new to this website and want to know much familiar infomration and real time examples what they can offer. 
"""
```

```python
user_prompt = """
    Now, I'll be adding the website and if articles are present. Go through it carefully and get the information correct 
    and provide knowldge accordingly.
"""
```

```python
messages = [

{"role" : "system", "content": "You are a very helpful assistant"},
{"role": "user", "content": "What is this website/article about ?"}
] # fill this in

response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
response.choices[0].message.content
```

```python
def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt+ website}
    ]
```

```python
messages_for(web)
```

```python
def summarize(url):
    website = fetch_website_contents(url)
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages_for(web))
    return response.choices[0].message.content
```

```python
summarize("https://www.walmart.com/")
```

```python
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
```
